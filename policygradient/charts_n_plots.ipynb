```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()

import glob
import pandas as pd
```

```python
files = glob.glob('.\\artefacts\\*policygradient.csv')
files
```

```python
fig,ax=plt.subplots(1,3,sharex='col',sharey= False,figsize=(16,4))
for n,file in enumerate(files):
    env_id = file.split('\\')[2].split('_')[0]
    df = pd.read_csv(file)
    df['mean_rewards'] = df['rewards'].rolling(50).mean()
    
    sns.lineplot(df['rewards'], ax = ax[n])
    sns.lineplot(df['mean_rewards'],ax = ax[n])
    ax[n].set_title(f'{env_id} Rewards & Mean Rewards',fontsize=8)
```
